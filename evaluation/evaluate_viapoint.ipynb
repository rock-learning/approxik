# Analysis of Viapoint Problem

```python
%pylab inline
import pickle
from scipy.stats import sem
```

```python
results = map(lambda f: pickle.load(open(f, "r")),
              ["results_benchmark_viapoint_approxik.pickle",
               "results_benchmark_viapoint_exactik.pickle",
               "results_benchmark_viapoint_joint.pickle",
               ])
methods = ["Approximate IK",
           "Exact IK",
           "Joint space"
           ]
```

```python
n_methods = len(results)
n_runs = len(results[0])
n_episodes = min([len(results[i][0][1]) for i in range(len(results))])


res = dict((method, np.empty((n_runs, n_episodes)))
           for method in methods)
runs = {method: 0 for method in methods}
for i in range(len(methods)):
    for j in range(n_runs):
        name, learning_curve, final = results[i][j]
        res[methods[i]][runs[methods[i]]] = learning_curve[:n_episodes]
        runs[methods[i]] += 1
```

```python
figure(figsize=(4, 3), dpi=200)
subplots_adjust(left=0.15, right=0.95, bottom=0.14, top=0.98)
colors = ["k", "k", "k"]
linstyles = ["-", "--", ":"]
step_size = 10
n_ep = 2000
x = np.arange(1, n_ep + 1, step_size)
for i, method in enumerate(methods):
    runs = res[method]
    m = np.maximum.accumulate(runs.T).T.mean(axis=0)[:n_ep:step_size]
    #s = np.maximum.accumulate(runs.T).T.std(axis=0)[:n_ep:step_size]
    s = sem(np.maximum.accumulate(runs.T).T, axis=0)[:n_ep:step_size]
    plot(x, m, label=method, lw=2, c=colors[i], ls=linstyles[i])
    # Uncomment to plot all runs
    #plot(x, runs[:, :n_ep:step_size].T)
    #plot(x, np.maximum.accumulate(runs.T)[:n_ep:step_size])
    fill_between(x, m - s, m + s, alpha=0.3, color=colors[i])
xlabel("Episode")
ylabel("Reward")
yscale("symlog")
ylim((-100, -5))
legend(loc="lower right")
savefig("result_comparison_viapoint.pdf")
```
